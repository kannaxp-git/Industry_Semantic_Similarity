```python
import numpy as np
import gensim
from scipy import spatial
import pandas as pd
import re
```

```python
w2v_model=gensim.models.KeyedVectors.load_word2vec_format(r'D:\workarea_kannan\word2vec\GoogleNews-vectors-negative300.bin', binary=True)
```

```python
def preProcess(txt):
    
    stopwordsKB=['unknown','other']
    
    txt=str(txt)
    txt=txt.lower()
    #Replacing Special Chars
    txt=txt.replace(' ','_')
    txt = re.sub('\W+','', txt ) # L&T --> LT Vs L T
    txt=txt.replace('_',' ')
  
    tokens=txt.split(' ')
    tokens=[t for t in tokens if t not in stopwordsKB]
    return tokens
```

```python

def avgFeatureVector(words, model=w2v_model, num_features=300):
        #function to average all words vectors in a given paragraph
        featureVec = np.zeros((num_features,), dtype="float32")
        nwords = 0

        #list containing names of words in the vocabulary
        index2word_set = set(model.index2word) #this is moved as input param for performance reasons
        for word in words:
            if word in index2word_set:
                nwords = nwords+1
                featureVec = np.add(featureVec, model[word])

        if(nwords>0):
            featureVec = np.divide(featureVec, nwords)
        return featureVec
```

```python
#get average vector for sentence 1
s1 = "Further/Continuing Education"
s1_avg_vector = avgFeatureVector(preProcess(s1), model=w2v_model, num_features=300)

#get average vector for sentence 2
s2 = "Private Universities"
s2_avg_vector = avgFeatureVector(preProcess(s2), model=w2v_model, num_features=300)

sen1_sen2_similarity =  1-spatial.distance.cosine(s1_avg_vector,s2_avg_vector)
print(sen1_sen2_similarity)
```
